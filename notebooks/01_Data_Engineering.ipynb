### **7: Define Column Names (Schema Engineering)**

```python
columns = (
    ["engine_id", "cycle"] +
    [f"op_setting_{i}" for i in range(1, 4)] +   # 3
    [f"sensor_{i}" for i in range(1, 24)]        # 23 sensors
)

len(columns)
```

```python
df.columns = columns
```

```python
df.head()
```

### **8: Separate Identifier vs Features**

```python
id_cols = ["engine_id", "cycle"]
feature_cols = [c for c in df.columns if c not in id_cols]
```

### **9: Understand the Missing Values**

```python
missing_ratio = df.isna().mean().sort_values(ascending=False)
missing_ratio[missing_ratio > 0]
```

```python
drop_cols = missing_ratio[missing_ratio > 0.9].index.tolist()
drop_cols
```

```python
df = df.drop(columns=drop_cols)
```

### **10: Save Cleaned Output**

```python
clean_path = "../data/processed/train_engine_data_clean.csv"
df.to_csv(clean_path, index=False)
```
